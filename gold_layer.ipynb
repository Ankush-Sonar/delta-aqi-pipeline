```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
from pyspark.sql import functions as F, Window


# Initialize an empty DataFrame with schema
schema = StructType([
    StructField("loc", StringType(), True),
    StructField("month", StringType(), True),
    StructField("avg_aqi", DoubleType(), True)
])
df_union = spark.createDataFrame([], schema)

# Create Table to see Monthly Average AQI of Each City - Monthly AQI
for tables in spark.catalog.listTables("aqi_cat.silver_schema"):
    df = spark.table("aqi_cat.silver_schema."+ tables.name)
    df_res=df.withColumn("month", date_format("date", "MM-yyyy",)).groupBy("loc", "month").agg(round(avg("aqi_pm25").alias("avg_aqi"),2))
    df_union = df_union.union(df_res)
display(df_union)
```
